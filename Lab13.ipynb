# W13A10

```python
def getMoney(path :str):
    with open(path, 'r') as f:
        a = list(map(int,f.readline().split()))
        if len(a) == 0:
            return 0
        if len(a) == 1:
            return a[0]
        n = len(a)
        dp = [0] * n
        dp[0] = a[0]
        dp[1] = max(a[0],a[1])
        for i in range(2,n):
            dp[i] = max(dp[i-1], dp[i-2] + a[i])
    return dp[n-1]
file = input()
print(getMoney(file))
```

# W13A14

```python
def averageTime(path: str):
    f = open(path, "r")
    lines = f.readlines()
    f.close()
    total = 0
    count = 0
    for line in lines[1:]:
        line = line.strip()
        if not line:
            continue
        parts = line.split(",")
        if len(parts) < 2:
            continue
        time = float(parts[1])
        total += time
        count += 1
    if count == 0:
        return 0
    return round(total / count, 2)
file = input()
print(averageTime(file))
```

# W13A15

```python
def findMovies(path: str):
    f = open(path, "r")
    lines = f.readlines()
    f.close()
    res = []
    for line in lines[1:]:
        parts = [x.strip() for x in line.split(",")]
        if len(parts) < 4:
            continue
        movie = parts[1]
        desc = parts[2].lower()
        rating = float(parts[3])
        if rating > 8.0 and "boring" not in desc:
            res.append(movie)
    if not res:
        return ["All are boring!"]
    return res
file = input()
print(findMovies(file))
```

# W13A16

```python
def shapesPerimeters(path: str):
    f = open(path, "r")
    lines = [x.strip() for x in f.readlines()]
    f.close()
    res = []
    i = 0
    while i < len(lines):
        if lines[i].startswith("TYPE"):
            shape = lines[i].split(":")[1].strip().upper()
            i += 1
            if shape == "SQUARE":
                side = float(lines[i].split(":")[1])
                res.append(f"{round(4 * side, 2):.2f}")
            elif shape == "RECTANGLE":
                parts = lines[i].replace(":", " ").split()
                w = float(parts[1])
                h = float(parts[3])
                res.append(f"{round(2 * (w + h), 2):.2f}")
            elif shape == "CIRCLE":
                r = float(lines[i].split(":")[1])
                res.append(f"{round(2 * 3.1415926535 * r, 2):.2f}")
        i += 1
    return res
file = input()
print(shapesPerimeters(file))
```

# W13A17

```python
def productExceptSelf(path: str):
    f = open(path, "r")
    nums = list(map(int, f.read().strip().split()))
    f.close()
    n = len(nums)
    res = [1] * n
    prefix = 1
    for i in range(n):
        res[i] = prefix
        prefix *= nums[i]
    suffix = 1
    for i in range(n - 1, -1, -1):
        res[i] *= suffix
        suffix *= nums[i]
    return res
file = input()
print(productExceptSelf(file))
```

# W13A18

```python
def estimatedTime(path: str):
    f = open(path, "r")
    lines = [x.strip() for x in f.readlines() if x.strip()]
    f.close()
    sx = sy = dx = dy = v = 0
    for line in lines:
        parts = line.replace(":", "").replace("KM/S", "").split()
        if parts[0] == "SRC":
            sx = float(parts[2])
            sy = float(parts[4])
        elif parts[0] == "DES":
            dx = float(parts[2])
            dy = float(parts[4])
        elif parts[0] == "VELOCITY":
            v = float(parts[1])
    d = ((dx - sx)**2 + (dy - sy)**2)**0.5
    t = d / v
    return round(t, 2)
file = input()
print(estimatedTime(file))
```

```python
def readDoubles(path: str):
    f = open(path, "rb")
    b = f.read()
    f.close()
    res = []
    size = 8
    for i in range(0, len(b), size):
        chunk = b[i:i+size]
        bits = int.from_bytes(chunk, "little")
        sign = -1 if (bits >> 63) else 1
        exp = ((bits >> 52) & 0x7FF) - 1023
        mant = bits & ((1 << 52) - 1)
        val = 1.0
        for j in range(52):
            if (mant >> (51 - j)) & 1:
                val += 2 ** -(j + 1)
        num = sign * val * (2 ** exp)
        res.append(f"{round(num, 5):.5f}")
    return res
file = input()
print(readDoubles(file))
```

```python
def imageSmoother(path: str):
    f = open(path, "r")
    lines = [x.strip() for x in f.readlines() if x.strip()]
    f.close()
    M = int(lines[0])
    mat = [list(map(int, lines[i + 1].split())) for i in range(M)]
    res = [[0] * M for _ in range(M)]
    for i in range(M):
        for j in range(M):
            s = 0
            c = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    x = i + dx
                    y = j + dy
                    if 0 <= x < M and 0 <= y < M:
                        s += mat[x][y]
                        c += 1
            res[i][j] = s // c
    return res
file = input()
print(imageSmoother(file))
```
