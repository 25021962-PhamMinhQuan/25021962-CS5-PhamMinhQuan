# W7A7

```python
def length(nums):
    lis = [] 
    for x in nums:
        left, right = 0, len(lis)
        while left < right:
            mid = (left + right) // 2
            if lis[mid] < x:
                left = mid + 1
            else:
                right = mid
        if left == len(lis):
            lis.append(x)
        else:
            lis[left] = x
    return len(lis)

a = list(map(int,input().split()))
print(length(a))
```

# W7A8

```python
def lower_bound(arr,x):
    l, r = 0, len(arr)
    while l < r:
        mid = (l + r)//2
        if arr[mid] < x:
            l = mid + 1
        else:
            r = mid
    return l
intervals = eval(input())
questions = list(map(int,input().split()))
ans = [-1]*len(questions)
for l, r in intervals:
    pos = lower_bound(questions,l)
    while pos < len(questions) and questions[pos] <= r:
        if ans[pos] == -1 or ans[pos] > r - l + 1:
            ans[pos] = r - l + 1
        pos += 1
print(ans)
```

# W7A9

```python
def binary_search(a, x):
    l, r = 0, len(a)
    while l < r:
        mid = (l+r)//2
        if x > a[mid]:
            l = mid + 1
        else:
            r = mid
    return l
def count_smaller(arr):
    sorted_list = []
    res = [0]*len(arr)
    for i in range(len(arr)-1,-1,-1):
        pos = binary_search(sorted_list, arr[i])
        res[i] = pos
        sorted_list.insert(pos,arr[i])
    return res
a = list(map(int,input().replace("[","").replace("]","").replace(",","").split()))
print(count_smaller(a))
```

# W7A10

```python
def max_mod(nums, m):
    ans = 0
    prefix = 0
    seen = [0]
    for x in nums:
        prefix = (prefix + x) % m
        ans = max(ans, prefix)
        l, r = 0,len(seen)
        while l < r:
            mid = (l + r) // 2
            if seen[mid] <= prefix:
                l = mid + 1
            else:
                r = mid
        if l < len(seen):
            ans = max(ans,(prefix - seen[l] + m) % m)
        seen.insert(l,prefix)
    return ans
a = list(map(int,input().split()))
m = int(input())
print(max_mod(a,m))
```
