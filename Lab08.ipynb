```python
class cylinder:
    def __init__(self, ban_kinh, chieu_cao):
        self.ban_kinh = ban_kinh
        self.chieu_cao = chieu_cao
    def Dtich(self):
        return (((self.ban_kinh) ** 2 ) * 3.14 * 2) + 2*self.ban_kinh*3.14*self.chieu_cao
    def Ttich(self):
        return ((self.ban_kinh)**2) * 3.14 * self.chieu_cao
r, h = map(float, input().split())
cyl = cylinder(r, h)

print(f"{cyl.Dtich():.2f}")
print(f"{cyl.Ttich():.2f}")
```

```python
class Date:
    def __init__(self, Ngay, Thang, Nam):
        self.Ngay = Ngay
        self.Thang = Thang
        self.Nam = Nam
    def exist(self):
        if self.Ngay < 0 or self.Thang < 0 or self.Nam < 0 or self.Thang>12:
            return "Invalid"
        else:
            if self.Thang in ( 1, 3, 5, 7, 8, 10, 12):
                if self.Ngay > 31:
                    return "Invalid"
                elif self.Ngay == 31:
                    if self.Thang != 12:
                        return f"1/{self.Thang+1}/{self.Nam}"
                    else:
                        return f"1/1/{self.Nam+1}"
                else:
                    return f"{self.Ngay+1}/{self.Thang}/{self.Nam}"
            elif self.Thang in (4, 6, 9, 11):
                if self.Ngay > 30:
                    return "Invalid"
                elif self.Ngay==30:
                    return f"1/{self.Thang+1}/{self.Nam}"
                else:
                    return f"{self.Ngay+1}/{self.Thang}/{self.Nam}"
            elif self.Thang == 2:
                if ((self.Nam % 4 == 0 and self.Nam % 100 != 0) or (self.Nam % 400 == 0)):
                    if self.Ngay > 29:
                        return "Invalid"
                    elif self.Ngay==29:
                        return f"1/3/{self.Nam}"
                    else:
                        return f"{self.Ngay+1}/2/{self.Nam}"
                elif self.Ngay > 28:
                    return "Invalid"
                elif self.Ngay == 28:
                    return f"1/3/{self.Nam}"
                else:
                    return f"{self.Ngay+1}/2/{self.Nam}"
            
d, m, y = map(int,input().split())
Ndate = Date(d,m,y)
print(Ndate.exist())
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add(self, name, price):
        self.items.append({"name": name, "price": int(price)})
        print("Da them")

    def remove(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                print("Da xoa san pham")
                return
        print("Khong tim thay san pham")

    def empty(self):
        if len(self.items) == 0:
            return "Empty"
        return "Not empty"

    def sum(self):
        ans = 0
        for item in self.items:
            ans += item["price"]
        return ans

    def Show(self):
        if self.empty() == "Empty":
            print("Gio hang rong")
        else:
            for item in self.items:
                print(item["name"], item["price"])

    def clear(self):
        self.items.clear()
        print("Da xoa toan bo gio hang")
cart = ShoppingCart()

while True:
    choice = input()

    if choice == "1":
        cart = ShoppingCart()
        print("Da khoi tao gio hang")

    elif choice == "2":
        Name, Price = input().split()
        cart.add(Name, Price)

    elif choice == "3":
        Name = input()
        cart.remove(Name)

    elif choice == "4":
        print(cart.empty())

    elif choice == "5":
        print(cart.sum())

    elif choice == "6":
        cart.Show()

    elif choice == "7":
        cart.clear()

    elif choice.lower() == "yes":
        break
```

```python
class Fraction:
    def __init__(self,x,y):
        if  y == 0:
            raise ValueError("Mau phai lon hon 0")
        elif y < 0:
            x = -x
            y = -y
        g = math.gcd(x,y)
        self.x = x//g
        self.y = y//g
    def __add__(self,other):
        x = self.x * other.y + self.y * other.x
        y = self.y * other.y
        return Fraction(x,y)
    def __sub__(self,other):
        x = self.x * other.y - self.y * other.x
        y = self.y * other.y
        return Fraction(x,y)
    def __mul__(self,other):
        x = self.x * other.x
        y = self.y * other.y
        return Fraction(x,y)
    def __truediv__(self,other):
        x = self.x * other.y
        y = self.y * other.x
        return Fraction(x,y)
    def __str__(self):
        return f"{self.x}/{self.y}"
a,b,op,c,d = input().split()
f1 = Fraction(int(a),int(b))
f2 = Fraction(int(c),int(d))
if op == "+":
    print(f1+f2)
elif op == "-":
    print(f1-f2)
elif op == "*":
    print(f1*f2)
elif op == "/":
    print(f1/f2)


        
```

```python
class BankAccount:
    def __init__(self, initial_balance):
        self.initial_balance = initial_balance
    def deposit(self, additional):
        return BankAccount(self.initial_balance + additional)
    def withdraw(self, lost):
        if self.initial_balance >= lost:
            return BankAccount(self.initial_balance - lost)
        else:
            return BankAccount(self.initial_balance)
    def __str__(self):
        return f"{self.initial_balance}"        
Name, Balance = input().split()
Command = int(input())
res = BankAccount(int(Balance))
for i in range(0,Command):
    choice, amount = input().split()
    amount = int(amount)
    if choice == "DEPOSIT":
        res = res.deposit(amount)
    elif choice == "WITHDRAW":
        res = res.withdraw(amount)
print(res)
```

```python
import math
class surface:
    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z
    def phaptuyen(self, other, another):
        vector1 = [self.x - other.x, self.y - other.y , self.z - other.z]
        vector2 = [self.x - another.x, self.y - another.y, self.z - another.z]
        return [vector1[1]*vector2[2] - vector2[1]*vector1[2], vector1[2]*vector2[0] - vector2[2]*vector1[0], vector1[0]*vector2[1] - vector2[0]*vector1[1]]
    def angle(self, other):
        return float((self.x * other.x + self.y * other.y + self.z * other.z)/(((self.x**2 + self.y**2 + self.z**2)**(1/2)) * (other.x**2 + other.y**2 + other.z**2)**(1/2)))

for i in range(4):
    x,y,z = map(float,input().split())
    if i == 0:
        A = surface(x,y,z)
    elif i == 1:
        B = surface(x,y,z)
    elif i ==2:
        C = surface(x,y,z)
    else: 
        D = surface(x,y,z)
cache1 = A.phaptuyen(B,C) 
phaptuyen1 = surface(cache1[0],cache1[1],cache1[2])
cache2 = B.phaptuyen(C,D)
phaptuyen2 = surface(cache2[0],cache2[1],cache2[2])
print(f"{math.degrees(math.acos(phaptuyen1.angle(phaptuyen2))):.2f}")
```

```python
class ComplexNumber:
    def __init__(self, initRealPart=0.0, initImaginaryPart=0.0):
        self.realPart = initRealPart
        self.imaginaryPart = initImaginaryPart

    def multiply(self, other):
        real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        imag = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(real, imag)
a, b = map(float, input().split())
c, d = map(float, input().split())
num1 = ComplexNumber(a, b)
num2 = ComplexNumber(c, d)
product = num1.multiply(num2)
print(product.realPart, product.imaginaryPart)
```

```python
import math
class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1 = initEdge1
        self.edge2 = initEdge2
        self.edge3 = initEdge3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        if a + b <= c or a + c <= b or b + c <= a:
            return -1
        p = (a + b + c) / 2
        area = math.sqrt(p * (p - a) * (p - b) * (p - c))
        return area
a, b, c = map(int, input().split())
triangle = Triangle(a, b, c)
area = triangle.getArea()

if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```
