# W4A11

```python
def uoc(n):
    if n % 2 != 0:
        return 0 
    n //= 2
    count = 0
    for i in range(1, int(n**(1/2)) + 1):
        if n % i == 0:
            count += 1
            if i != n // i:
                count += 1
    return count

n = int(input())
print(uoc(n))
```

# W4A26

```python
def largest_fib(n):
	if n<1:
		return 0
	f1,f2=1,1
	while f2<=n:
		f1 , f2 =f2 , f2 + f1
	return f1
n=int(input())
print(largest_fib(n))
```
